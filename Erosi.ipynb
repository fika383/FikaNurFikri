```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def proses_erosi(file_gambar):
    # Membaca gambar dalam grayscale
    citra = cv2.imread(file_gambar, 0)

    # Cek apakah gambar berhasil dibaca
    if citra is None:
        print(f"Gagal membaca gambar: {file_gambar}")
        return

    # Membuat struktur elemen (kernel)
    elemen_struktural = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))

    # Menerapkan operasi erosi
    citra_erosi = cv2.erode(citra, elemen_struktural, iterations=1)

    # Menampilkan citra asli dan hasil erosi
    tampilkan_visual(citra, citra_erosi, file_gambar)

def tampilkan_visual(citra_awal, citra_erosi, judul_file):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    fig.suptitle(f'Proses Erosi pada {judul_file}', fontsize=13)

    axes[0].imshow(citra_awal, cmap='gray')
    axes[0].set_title('Citra Asli')
    axes[0].axis('off')

    axes[1].imshow(citra_erosi, cmap='gray')
    axes[1].set_title('Citra Setelah Erosi')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

# Jalankan proses erosi untuk kedua gambar
proses_erosi("1.png")
proses_erosi("2.png")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def proses_opening(file_gambar):
    # Membaca gambar dalam mode grayscale
    citra = cv2.imread(file_gambar, 0)

    # Cek apakah gambar berhasil dibaca
    if citra is None:
        print(f"Gagal membaca gambar: {file_gambar}")
        return

    # Membuat struktur elemen (kernel)
    elemen_struktural = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))

    # Menerapkan operasi opening: erosi lalu dilasi
    citra_opening = cv2.morphologyEx(citra, cv2.MORPH_OPEN, elemen_struktural)

    # Menampilkan citra asli dan hasil opening
    tampilkan_visual(citra, citra_opening, file_gambar)

def tampilkan_visual(citra_awal, citra_opening, judul_file):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    fig.suptitle(f'Proses Opening pada {judul_file}', fontsize=13)

    axes[0].imshow(citra_awal, cmap='gray')
    axes[0].set_title('Citra Asli')
    axes[0].axis('off')

    axes[1].imshow(citra_opening, cmap='gray')
    axes[1].set_title('Citra Setelah Opening')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

# Jalankan proses opening untuk kedua gambar
proses_opening("1.png")
proses_opening("2.png")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def proses_dilasi(file_gambar):
    # Membaca gambar dalam grayscale
    citra = cv2.imread(file_gambar, 0)

    # Membuat struktur elemen (kernel)
    elemen_struktural = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))

    # Menerapkan operasi dilasi
    citra_dilasi = cv2.dilate(citra, elemen_struktural, iterations=1)

    # Menampilkan citra asli dan hasil dilasi
    tampilkan_visual(citra, citra_dilasi, file_gambar)

def tampilkan_visual(citra_awal, citra_dilasi, judul_file):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    fig.suptitle(f'Proses Dilasi pada {judul_file}', fontsize=13)

    axes[0].imshow(citra_awal, cmap='gray')
    axes[0].set_title('Citra Asli')
    axes[0].axis('off')

    axes[1].imshow(citra_dilasi, cmap='gray')
    axes[1].set_title('Citra Setelah Dilasi')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

# Jalankan proses untuk kedua gambar
proses_dilasi("1.png")
proses_dilasi("2.png")
```
